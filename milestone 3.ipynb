# 4. Model selection and evaluation


```python
#loading the scaler
from sklearn.preprocessing import StandardScaler
cat_scaler = StandardScaler()
txn_scaler = StandardScaler()
Gender_scaler = StandardScaler()
zip_scaler = StandardScaler()
Job_scaler = StandardScaler()
trans_hour_scaler = StandardScaler()
cust_age_groups_scaler = StandardScaler()
amount_to_avg_ratio_scaler = StandardScaler()
```

```python
X_train['Merchant_category'] = cat_scaler.fit_transform(X_train[['Merchant_category']])
X_train['Transaction_amount'] = txn_scaler.fit_transform(X_train[['Transaction_amount']])
X_train['Gender'] = Gender_scaler.fit_transform(X_train[['Gender']])
X_train['zip'] = zip_scaler.fit_transform(X_train[['zip']])
X_train['Job'] = Job_scaler.fit_transform(X_train[['Job']])
X_train['trans_hour'] = trans_hour_scaler.fit_transform(X_train[['trans_hour']])
X_train['cust_age_groups'] = cust_age_groups_scaler.fit_transform(X_train[['cust_age_groups']])
X_train['amount_to_avg_ratio'] = amount_to_avg_ratio_scaler.fit_transform(X_train[['amount_to_avg_ratio']])
```

```python
X_test['Merchant_category'] = cat_scaler.transform(X_test[['Merchant_category']])
X_test['Transaction_amount'] = txn_scaler.transform(X_test[['Transaction_amount']])
X_test['Gender'] = Gender_scaler.transform(X_test[['Gender']])
X_test['zip'] = zip_scaler.transform(X_test[['zip']])
X_test['Job'] = Job_scaler.transform(X_test[['Job']])
X_test['trans_hour'] = trans_hour_scaler.transform(X_test[['trans_hour']])
X_test['cust_age_groups'] = cust_age_groups_scaler.transform(X_test[['cust_age_groups']])
X_test['amount_to_avg_ratio'] = amount_to_avg_ratio_scaler.transform(X_test[['amount_to_avg_ratio']])
```

```python
ModelName=[]
ModelAccuracy=[]
for name, model in tqdm(models.items()):
    model.fit(X_train, y_train)
    ModelName.append(name)
    y_pred = model.predict(X_test)
    ModelAccuracy.append([
         accuracy_score(y_test, y_pred),
         precision_score(y_test, y_pred),
         recall_score(y_test, y_pred),
         f1_score(y_test,y_pred)
         ])
ModelAccuracy=pd.DataFrame(ModelAccuracy,index=ModelName,columns=['Accuracy','Precision','Recall','F1 Score'])
ModelAccuracy
```

# Random Forest Classifier

```python
rf = RandomForestClassifier(class_weight = 'balanced')
```

```python
# fitting on the training data
rf.fit(X_train,y_train)

# predictions on the training data
y_pred_train_rf = rf.predict(X_train)

# predictions on the test data
y_pred_test_rf = rf.predict(X_test)
```

```python
# check confusion matrix on both train and test data

## TN FP
## FN TP
cm_train = confusion_matrix(y_train,y_pred_train_rf)
cm_test = confusion_matrix(y_test,y_pred_test_rf)
print('Training data:',cm_train)
print()
print('Testing data:', cm_test)
```

```python
plt.figure(figsize=(12, 5))

# Confusion Matrix - Train
plt.subplot(1, 2, 1)
sns.heatmap(cm_train, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No Anemia", "Anemia"],
            yticklabels=["No Anemia", "Anemia"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Train Confusion Matrix")

# Confusion Matrix - Test
plt.subplot(1, 2, 2)
sns.heatmap(cm_test, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No Anemia", "Anemia"],
            yticklabels=["No Anemia", "Anemia"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Test Confusion Matrix")

plt.tight_layout()
plt.show()
```

```python
# check recall scores on test data
print(f'Recall score of the model on test data: {cm_test[1][1]/(cm_test[1][1]+cm_test[1][0])}')
```

```python
# get the parameters of the model
rf.get_params()
```

```python
rf = RandomForestClassifier(class_weight = 'balanced')
params = {'n_estimators':[5,10,20,40,80,100,200],
          'criterion':['gini','entropy','log_loss'],
          'max_depth':[2,5,10,20],
          'min_samples_split':[2,4,8,12],
          'min_samples_leaf':[2,4,6,8,10],
          'oob_score': [True]}
#randomized search cv
rcv = RandomizedSearchCV(estimator = rf, param_distributions = params, n_iter = 10, scoring = 'recall', cv = 10, verbose=5, 
                                random_state=42, n_jobs=-1, return_train_score=True)
rcv.fit(X_train,y_train)
```

```python
# best estimator
rcv.best_estimator_
```

```python
# best score
rcv.best_score_
```

```python
# let's re build random forest model
rf = rcv.best_estimator_

# fitting on train data
rf.fit(X_train,y_train)
```

```python
# predictions on train data
y_train_pred_rf = rf.predict(X_train)

# predictions on test data
y_test_pred_rf = rf.predict(X_test)
```

```python
# check confusion matrix on both train and test data
## TN FP
## FN TP
cm_train = confusion_matrix(y_train,y_train_pred_rf)
cm_test = confusion_matrix(y_test,y_test_pred_rf)
print('Training data:',cm_train)
print()
print('Testing data:', cm_test)
```

```python
# check recall scores on train data
print(f'Recall score of the model on train data: {cm_train[1][1]/(cm_train[1][1]+cm_train[1][0])}')
# check recall scores on test data
print(f'Recall score of the model on test data: {cm_test[1][1]/(cm_test[1][1]+cm_test[1][0])}')
```
